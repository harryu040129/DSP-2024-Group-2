```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 불러오기
file_path = 'color - Sheet1 (2).csv'  # 파일 이름을 실제 업로드된 이름으로 수정
data = pd.read_csv(file_path)

# 데이터 미리보기
print("데이터 프레임 미리보기:")
print(data.head())

# 1. 전체 선호도를 기준으로 정렬
top_colors = data.sort_values(by='Overall_Preference(%)', ascending=False)

# 정렬된 데이터 출력
print("\n가장 선호되는 색상 순위:")
print(top_colors[['Color', 'Overall_Preference(%)']])

# 2. 전체 선호도 기준 색상 막대 그래프
plt.figure(figsize=(10, 6))
plt.bar(top_colors['Color'], top_colors['Overall_Preference(%)'], color='skyblue')
plt.xlabel('Color')
plt.ylabel('Preference (%)')
plt.title('Overall Color Preference')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
